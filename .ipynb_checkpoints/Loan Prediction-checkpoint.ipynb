# Loan Prediction System

This system is used to make predictions for loan approval of target customers in a private company on the basis of various parameters/features. The prediction is calculated from the data obtained from Kaggle. It is a classification system which makes use of Machine Learining algorithms for the predict operation. * To find the relation between the features from the numerical values, one way is to calculate the variance inflation factor(VIF) of the dataset ranging from 0 to 10. 
* The VIF values less thn 5 are not linearly related while, values between 6 to 10 have chances of correlation.
* In order to reduce VIF value, one can remove those features to maintain the values but, this does not indicate the importance of the features.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = loan_data_new.drop(['Loan_Status'], axis=1)
vif["variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(0,X.shape[1])]

vif
```

### 4. Column Standardization


* Before jumping directly to modeling, we first need to prepare our dataset to be ready for modeling which requires column standardization.
* Column standardization is a technique to scale the dataset into a range(mostly 0 to 1) as outliers will not be affected by standardization and process and easy to find them.
* In this scaling technique, the values are centered around with mean having a unit standard deviation(mean=0, sd=1).

                             stand_X = (X - Feature scaling: Mu) / Feature scaling: Sigma
                             
* Here, Mu is the mean and Sigma is the standard deviation.                       

```python
#Standardize data for prediction using StandardScaler
from sklearn.preprocessing import StandardScaler

X = loan_data_new.drop(['Loan_Status'], axis=1)
print(X)
scaler = StandardScaler()
stand_X = scaler.fit_transform(X)
print(stand_X)
```

```python
y = loan_data_new['Loan_Status']
y
```

* All the features are stored in variable X and after the column standardization, they are stored in stand_X variable.

### 5. Modeling



```python
#Splitting data into train and test
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(stand_X, y, test_size=0.2, random_state=42)

print("train data ", train_X.shape, train_y.shape)
print("test data ", test_X.shape, test_y.shape)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(penalty='l1', C=1.0, max_iter=100, solver='liblinear')
LR.fit(train_X, train_y)
```

```python
y_pred = LR.predict(test_X)

y_pred
```

```python
from sklearn.metrics import confusion_matrix

arr = confusion_matrix(test_y, y_pred)
print(arr)
```

```python
f, ax = plt.subplots(figsize=(6, 4))
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in arr.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     arr.flatten()/np.sum(arr)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(arr, annot=labels, fmt='', cmap="Blues")
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(test_y, y_pred)
print(score)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(LR, train_X, train_y.ravel(), cv=10, scoring='accuracy')
avg_score = scores.mean()

print("\nAverage Cross-Validation Score: %f" %avg_score)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

cv_scores = []

#perform 10-fold cross-validation
for k in range(1,21):
    neigh = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(neigh, train_X, train_y.ravel(), cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

#Misclassification Error
MSE = [1- x for x in cv_scores]

#Optimal Value of K for KNN
best_k = MSE.index(min(MSE)) + 1
print("Best Value of K: ", best_k)
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(train_X, train_y.ravel())
```

```python
y_pred = knn.predict(test_X)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix

arr = confusion_matrix(test_y, y_pred)
print(arr)
```

```python
f, ax = plt.subplots(figsize=(6, 4))
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in arr.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     arr.flatten()/np.sum(arr)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(arr, annot=labels, fmt='', cmap="Blues")
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(test_y, y_pred)
print(score)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn, train_X, train_y.ravel(), cv=10, scoring='accuracy')
avg_score = scores.mean()

print("Average Cross-Validation Score: %f" %avg_score)
```
